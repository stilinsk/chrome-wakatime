```python
import matplotlib .pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df.plot(x='Gender',y='Age')
plt.ylabel('Age');
plt.xlabel('no of workers')
plt.title('Workers in the model')
```

```python
sns.countplot(data = df,x ='Age')
plt.xticks(rotation=90);
```

```python
sns.countplot(data=df,x='Main language at work')
plt.xticks(rotation=90);
```

##### According to the graph above English language is the most common language in the workplace

```python
city_age=df.groupby("City").Age.mean().reset_index().sort_values(by='Age',ascending=False)
sns.barplot(data=city_age[:20],x='City',y='Age')
plt.xticks(rotation=90);
```

```python
sns.set_style('whitegrid')
sns.countplot(data=df, x='City')
plt.title('Using seaborn')
plt.xticks(rotation=90);
```

```python
sns.set_style('darkgrid')
df.plot('City')
plt.title('Using seaborn')
```
